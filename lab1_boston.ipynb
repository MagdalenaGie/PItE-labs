```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression

boston = load_boston()
boston
```

```python
df = pd.DataFrame(boston.data, columns = boston.feature_names)
df['PRICE'] = pd.Series(boston.target)
df.head()
```

```python
histograms = df.hist(column = ['CRIM', 'AGE'])
```

```python
#trying different bins 
histograms = df.hist(column = ['CRIM', 'AGE'], bins=20)
```

```python
histograms = df.hist(column = ['CRIM', 'AGE'], bins=5)
```

```python
reg = LinearRegression()

for ind in df.columns:
    OX = df[ind].values.reshape(-1, 1)
    OY = boston.target
    reg.fit(OX, OY)

    ox = np.linspace(OX.min(), OX.max())
    oy = reg.intercept_ + reg.coef_*ox
    
    plt.title(ind)
    plt.scatter(OX, OY, c='y')
    plt.plot(ox, oy, 'b--')
    plt.show()
    
```

```python
print("Short interpretation of the result:")
print(" - there is a strong linear dependency for PRICE")
print(" - there is also some light linear dependency for INDUS, RM, AGE, PTRATIO and a little bit stronger one for LSTAT")
print(" - for factors like B, RAD, CHAS there is no linear dependency")
```
